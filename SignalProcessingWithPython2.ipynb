```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100
Fs = 1000
Ts = 1/Fs
a1, a2, a3 = 220, 20, 20
f1, f2, f3 = 50, 100, 400
p1, p2, p3 = np.deg2rad(0), np.deg2rad(30), np.deg2rad(45)
```

```python
t = np.arange(0,N*Ts,Ts)
```

```python
xt1 = a1*np.sin((2*np.pi*f1*t)+p1)
xt2 = a2*np.sin((2*np.pi*f2*t)+p2)
xt3 = a3*np.sin((2*np.pi*f3*t)+p3)
```

```python
plt.plot(t,xt1,t,xt2,t,xt3)
```

```python
plt.stem(t,xt1)
plt.stem(t,xt2)
plt.stem(t,xt3)
```

```python
xt = xt1+xt2+xt3
plt.plot(t,xt)
plt.show()
plt.stem(t,xt)
```

```python
spectrum = np.fft.fft(xt)
print(spectrum)
```
